# Price Trend Analysis â€“ Dairy Products

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned dataset
df = pd.read_csv("../data/dairy_prices_cleaned.csv", parse_dates=['Date'])
```

```python
# Set plot style
sns.set(style="whitegrid")
```

## 1. Average price over time (All countries/products)

```python

monthly_avg = df.groupby("Date")["Price_EUR_per_kg"].mean().reset_index()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_avg, x="Date", y="Price_EUR_per_kg")
plt.title("Average Dairy Price Over Time")
plt.ylabel("EUR/kg")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

## 2. Price trends by product (all countries)

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x="Date", y="Price_EUR_per_kg", hue="Product")
plt.title("Price Trends by Product")
plt.ylabel("EUR/kg")
plt.xlabel("Date")
plt.legend(title="Product")
plt.tight_layout()
plt.show()
```

## 3. Country-specific trend for one product

```python
milk_df = df[df["Product"] == "Whole Milk"]

plt.figure(figsize=(12, 5))
sns.lineplot(data=milk_df, x="Date", y="Price_EUR_per_kg", hue="Country")
plt.title("Whole Milk Prices by Country")
plt.ylabel("EUR/kg")
plt.xlabel("Date")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```
