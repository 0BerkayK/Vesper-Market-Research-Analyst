# Supply-Demand Balance Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv("../data/production_import_export.csv", parse_dates=["Date"])
```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

## 1. Calculate Net Supply: Production + Import - Export

```python

df["Net_Supply"] = df["Production_tons"] + df["Import_tons"] - df["Export_tons"]
```

## 2. Trend of Net Supply Over Time by Country

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Date", y="Net_Supply", hue="Country")
plt.title("Net Dairy Supply Over Time")
plt.ylabel("Net Supply")
plt.xlabel("Date")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```

## 3. Compare Supply Components

### Germany

```python
country = "Germany"
country_df = df[df["Country"] == country]

plt.figure(figsize=(12, 6))
plt.plot(country_df["Date"], country_df["Production_tons"], label="Production", marker="o")
plt.plot(country_df["Date"], country_df["Import_tons"], label="Import", marker="o")
plt.plot(country_df["Date"], country_df["Export_tons"], label="Export", marker="o")
plt.title(f"Supply Components in {country}")
plt.ylabel("Volume")
plt.xlabel("Date")
plt.legend()
plt.tight_layout()
plt.show()

price_df = pd.read_csv("../data/dairy_prices_cleaned.csv", parse_dates=["Date"])
price_country = price_df[price_df["Country"] == country].groupby("Date")["Price_EUR_per_kg"].mean().reset_index()

merged = pd.merge(country_df, price_country, on="Date")

plt.figure(figsize=(12, 6))
sns.scatterplot(data=merged, x="Net_Supply", y="Price_EUR_per_kg")
sns.regplot(data=merged, x="Net_Supply", y="Price_EUR_per_kg", scatter=False, color="red")
plt.title(f"Price vs Net Supply in {country}")
plt.xlabel("Net Supply")
plt.ylabel("Price (EUR/kg)")
plt.tight_layout()
plt.show()
```
